```python
import pandas as pd
data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

df = pd.DataFrame(data)
df.to_csv('employees.csv', index=False)
df.head()
shape = df.shape
print(f"Shape of the DataFrame: {shape}")
summary = df.info()
print("Summary of the DataFrame:\n", summary)
descriptive_stats = df.describe()
print("Descriptive Statistics:\n", descriptive_stats)
first_5 = df.head()
last_3 = df.tail(3)
print("First 5 rows:\n", first_5)
print("Last 3 rows:\n", last_3)
average_salary = df['Salary'].mean()
print(f"Average salary: {average_salary}")

total_bonus = df['Bonus'].sum()
print(f"Total bonus paid: {total_bonus}")

youngest_age = df['Age'].min()
print(f"Youngest employee's age: {youngest_age}")

highest_rating = df['Rating'].max()
print(f"Highest performance rating: {highest_rating}")
sorted_df = df.sort_values(by='Salary', ascending=False)

print("Data sorted by salary (descending):\n", sorted_df)
def performance_category(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

df['Performance_Category'] = df['Rating'].apply(performance_category)
print("Dataset with Performance Category:\n", df)
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print("Dataset with renamed column:\n", df)

more_than_5_years = df[df['Years_of_Experience'] > 5]
print("Employees with more than 5 years of experience:\n", more_than_5_years)

it_department = df[df['Department'] == 'IT']
print("Employees in the IT department:\n", it_department)

df['Tax'] = df['Salary'] * 0.1
print("Dataset with Tax column:\n", df)

df.to_csv('modified_employees.csv', index=False)
print("Modified dataset saved as 'modified_employees.csv'")



```
