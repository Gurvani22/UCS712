```python
import numpy as np
initials=['G','A']
X=sum(ord(initial) for initial in initials)
sales=np.array([X,X+50,X+100,X+150,X+200])
tax_rate = ((X % 5) + 5) / 100
sales_withtax=sales*(1+tax_rate)
discounted_sales = np.copy(sales)
discounted_sales[sales<X+100]=sales[sales<X+100]*0.95
discounted_sales[sales>=X+100]=sales[sales>=X+100]*0.90
sales_matrix=np.repeat([sales],3,axis=0)#np.repeat(a,repeats,axis=)
week_numbers = np.arange(1, 4)  # Create an array [1, 2, 3] to represent the weeks
sales_matrix = sales_matrix * (1 + 0.02) ** week_numbers[:, None]
# week_numbers[:, None] is a way of reshaping ke hum jo array hai usko coloumn mein krdein
print("sum of ascii values:",X)
print("sales datsaet",sales)
print("tax with sales are:",sales_withtax)
print("discounted sales are:",discounted_sales)
print("sales_matrix:",sales_matrix)




```

```python
import numpy as np
import matplotlib.pyplot as plt
#x=np.linspace(start,stop,num);
x=np.linspace(-10,10,100)
y1=x**2 #x**2=x^2
y2 = np.sin(x)  # Y = sin(x)
y3 = np.exp(x)  # Y = e^x
y4 = np.log(np.abs(x) + 1) # Y = log(|x| + 1)
#plt.subplot(nrows, ncols, index)
plt.subplot(2, 2, 1)  # 2x2 grid, 1st subplot
plt.plot(x, y1, label="Y = x^2", color="blue")
plt.title("Y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(2, 2, 2)  # 2x2 grid, 2st subplot
plt.plot(x, y1, label="Y = sin(x)", color="red")
plt.title("Y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(2, 2, 3)  # 2x2 grid, 3st subplot
plt.plot(x, y1, label="Y = exp(x)", color="green")
plt.title("Y = exp(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(2, 2, 4)  # 2x2 grid, 4st subplot
plt.plot(x, y1, label="Y =log(|x| + 1)", color="pink")
plt.title("Y = log(|x| + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)





```
